```python
import youtube_websearch as yt
import webbrowser
import pandas as pd
import numpy as np

def get_youtube_details(search_term = "13 Reasons Why"):
    search_term = search_term+" trailer"
    yt_basis = "https://www.youtube.com/watch?v="

    search_results = yt.plain_search(search_term)
    # search_results = yt.search('we will rock you', useragent="something")  # you can set your own useragent
    len(search_results)
    # videoId = search_results[0]["videoId"]

    # webbrowser.open(yt_basis + videoId)
    title = []
    views = []
    vid_len = []
    for i in range(len(search_results)):
        # title
        temp_title = search_results["contents"]["twoColumnSearchResultsRenderer"]["primaryContents"]["sectionListRenderer"]["contents"][0]["itemSectionRenderer"]["contents"][i]["videoRenderer"]["title"]["runs"][0]["text"]
        if "trailer" in temp_title or "teaser" in temp_title or "Trailer" in temp_title or "Teaser" in temp_title:
            title.append(temp_title)
        else:
            continue

        # views
        temp_views = search_results["contents"]["twoColumnSearchResultsRenderer"]["primaryContents"]["sectionListRenderer"]["contents"][0]["itemSectionRenderer"]["contents"][i]["videoRenderer"]["viewCountText"]["simpleText"]
        views.append(int((temp_views.split()[0]).replace(",", "")))

        # vid length
        vid_len.append(search_results["contents"]["twoColumnSearchResultsRenderer"]["primaryContents"]["sectionListRenderer"]["contents"][0]["itemSectionRenderer"]["contents"][i]["videoRenderer"]["lengthText"]["simpleText"])

    indx = views.index(max(views))
    return title[indx], views[indx], vid_len[indx]
```

```python
import youtube_websearch as yt
import webbrowser

def get_youtube_details(search_term="13 Reasons Why"):
    search_term = search_term + " trailer"
    
    yt_basis = "https://www.youtube.com/watch?v="
    search_results = yt.plain_search(search_term)
    
    # To get the list of Youtube video titles
    text_list = str(search_results).split("text': ")
    title = []
    for i in range(len(text_list)):
        temp = text_list[i].split("}")[0]
        if "railer" in temp and ("{" not in temp and "}" not in temp):
            title.append(temp.replace("'", ""))
            
    # To get their respective views and video duration
    text_list = str(search_results).split("simpleText': ")
    duration = []
    views = []
    for i in range(len(text_list)):
        temp = text_list[i].split("}")[0]
        if "{" not in temp and "}" not in temp:
            if ":" in temp:
                duration.append(temp.replace("'", ""))
            elif "," in temp and "views" in temp:
                views.append(int(((temp.split()[0]).replace("'", "")).replace(",", "")))

        if "People also watched" in temp:
            break
    indx = views.index(max(views))
    
    return (title[indx], duration[indx], views[indx])

def isNaN(num):
    return num != num
```

```python
df = pd.read_excel("total_data.xlsx")

for indx in df.index:
    try:
        (title, duration, views) = get_youtube_details(df["key"][indx])
#         (title, duration, views) = result
#         print(result)
        df["yt_title"][indx] = title
        df["yt_duration"][indx] = duration
        df["yt_views"][indx] = views
    except:
        df["yt_title"][indx] = ""
        df["yt_duration"][indx] = ""
        df["yt_views"][indx] = ""
        
        continue

df.to_excel("output-3.xlsx", index=False)
# for indx in df.index:
#     if isNaN(df["Number of YouTube views"][indx]):
#         try:
#             df2["Number of YouTube views"][indx] = get_youtube_details_new(df2["key"][indx])[1]
#         except:
#             continue
# df2.to_excel("output-2.xlsx", index=False)
```
