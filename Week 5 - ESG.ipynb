```python
def populate_years_and_values(xls, sheet_name, term):
    sheet = pd.read_excel(xls, sheet_name)
    for i in sheet.index:
        sheet_resource = []
        if term in str(sheet[sheet_name][i]):
            sheet_year = sheet.loc[0,].tolist()
            sheet_resource = sheet.loc[i,].tolist()
            years = get_years(sheet_year)
            values = get_values(sheet_resource)
            df = pd.DataFrame(list(zip(years, values)), columns =['Year', 'Value'])
            break
    return df
```

```python
def is_number(string):
    try:
        float(string)
        return True
    except ValueError:
        return False
```

```python
def get_average_closest(d, target):
    years = []
    for k,v in d.items():
        years.append(k)
    if not years:
        return 0
    elif(len(years) == 1):
        return d[years[0]]
    elif(len(years) == 2):
        return (d[years[0]] + d[years[1]])/2
    elif(years[0] < target):
            return (d[years[0]] + d[years[1]])/2
    elif(years[-1] > target):
            return (d[years[-1]] + d[years[-2]])/2
    else:
        val1 = min(years, key=lambda x:abs(x-target))
        years.remove(val1)
        val2 = min(years, key=lambda x:abs(x-target))
        return (d[val1] + d[val2])/2
        
```

```python
# env_resource[1:5]
# env_year[1:5]
def get_years(sheet_year):
    years = list(dict.fromkeys(sheet_year))
    years = [item for item in years if str(item) != 'nan']
    return years

def get_values(sheet_resource):
    values = []
    for i in range(1, len(sheet_resource)): 
        if i % 2 == 0: 
            values.append(sheet_resource[i])
            

#     values = [item if str(item) != 'nan' if '--' in str(item) '0' for item in values ]
    values = [item if str(item) != 'nan' else '0' for item in values]
    values = [str(sub).replace('--', '0') for sub in values] 
    return values
```

```python
from numbers import Number
def populate_df_for_csv(company, year):
    val = [0]*11
    year = str(year)
    try:
        xls = pd.ExcelFile('Firms ESG/' + company + '.xlsx')
    except FileNotFoundError:
        return val

    #env worksheet
    resource = populate_years_and_values(xls, 'Environment', "Resource Use")
    emission = populate_years_and_values(xls, 'Environment', "Emissions")
    innovation = populate_years_and_values(xls, 'Environment', "Innovation")

    #soc worksheet
    workforce = populate_years_and_values(xls, 'Social', "Workforce")
    human_rights = populate_years_and_values(xls, 'Social', "Human Rights")
    community = populate_years_and_values(xls, 'Social', "Community")
    product_responsibility = populate_years_and_values(xls, 'Social', "Product Responsibility")

    #gov worksheet
    management = populate_years_and_values(xls, 'Governance', "Management")
    shareholders = populate_years_and_values(xls, 'Governance', "Shareholders")
    csr_strategy = populate_years_and_values(xls, 'Governance', "CSR Strategy")


    x = 0
    resourceVal = {}
    emissionVal = {}
    innovationVal = {}
    workforceVal = {}
    human_rightsVal = {}
    communityVal = {}
    product_responsibilityVal = {}
    managementVal = {}
    shareholdersVal = {}
    csr_strategyVal = {}
    
    for i in resource['Year'].index:
        if year in str(resource['Year'][i]):
            val[x] = (str(resource['Value'][i]))
#         print(val[x])
        if(is_number(resource['Value'][i]) and resource['Value'][i] != '0'):
            resourceVal[resource['Year'][i]] = float(resource['Value'][i])

    x += 1
    for i in emission['Year'].index:
        if year in str(emission['Year'][i]):
            val[x] = (str(emission['Value'][i]))
#         print(val[x])
        if(is_number(emission['Value'][i]) and emission['Value'][i] != '0'):
            emissionVal[emission['Year'][i]] = float(emission['Value'][i])

    x += 1
    for i in innovation['Year'].index:
        if year in str(innovation['Year'][i]):
            val[x] = (str(innovation['Value'][i]))
        if(is_number(innovation['Value'][i]) and innovation['Value'][i] != '0'):
            innovationVal[innovation['Year'][i]] =float(innovation['Value'][i])

    x += 1
    for i in workforce['Year'].index:
        if year in str(workforce['Year'][i]):
            val[x] = (str(workforce['Value'][i]))
#         print(val[x])
        if(is_number(workforce['Value'][i]) and workforce['Value'][i] != '0'):
            workforceVal[workforce['Year'][i]] = float(workforce['Value'][i])

    x += 1
    for i in human_rights['Year'].index:
        if year in str(human_rights['Year'][i]):
            val[x] = (str(human_rights['Value'][i]))
#         print(val[x])
        if(is_number(human_rights['Value'][i]) and human_rights['Value'][i] !='0'):
            human_rightsVal[human_rights['Year'][i]] = float(human_rights['Value'][i])

    x += 1
    for i in community['Year'].index:
        if year in str(community['Year'][i]):
            val[x] = (str(community['Value'][i]))
#         print(val[x])
        if(is_number(community['Value'][i]) and community['Value'][i] !='0'):
            communityVal[community['Year'][i]] = float(community['Value'][i])

    x += 1
    for i in product_responsibility['Year'].index:
        if year in str(product_responsibility['Year'][i]):
            val[x] = (str(product_responsibility['Value'][i]))
#         print(val[x])
        if(is_number(product_responsibility['Value'][i]) and product_responsibility['Value'][i] != '0'):
            product_responsibilityVal[product_responsibility['Year'][i]] = float(product_responsibility['Value'][i])

    x += 1
    for i in management['Year'].index:
        if year in str(management['Year'][i]):
            val[x] = (str(management['Value'][i]))
#         print(val[x])
        if(is_number(management['Value'][i]) and management['Value'][i] != '0'):
            managementVal[management['Year'][i]] = float(management['Value'][i])

    x += 1
    for i in shareholders['Year'].index:
        if year in str(shareholders['Year'][i]):
            val[x] = (str(shareholders['Value'][i]))
#         print(val[x])
        if(is_number(shareholders['Value'][i]) and shareholders['Value'][i] != '0'):
            shareholdersVal[shareholders['Year'][i]] = float(shareholders['Value'][i])

    x += 1
    for i in csr_strategy['Year'].index:
        if year in str(csr_strategy['Year'][i]):
            val[x] = (str(csr_strategy['Value'][i]))
#         print(val[x])
        if(is_number(csr_strategy['Value'][i]) and csr_strategy['Value'][i] != '0'):
            csr_strategyVal[csr_strategy['Year'][i]] = float(csr_strategy['Value'][i])
            
    x += 1
    val[x] = '0'
            
    
    if(all(v == '0' for v in val)):
        val[0] = get_average_closest(resourceVal, int(year))
        val[1] = get_average_closest(emissionVal, int(year))
        val[2] = get_average_closest(innovationVal, int(year))
        val[3] = get_average_closest(workforceVal, int(year))
        val[4] = get_average_closest(human_rightsVal, int(year))
        val[5] = get_average_closest(communityVal, int(year))
        val[6] = get_average_closest(product_responsibilityVal, int(year))
        val[7] = get_average_closest(managementVal, int(year))
        val[8] = get_average_closest(shareholdersVal, int(year))
        val[9] = get_average_closest(csr_strategyVal, int(year))
        val[10] = '1'
            

    return val
```
