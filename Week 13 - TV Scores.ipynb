```python
def get_wiki_url(tv_name = "Hemlock Grove"):
    #Wikipedia
    # tv_name = "Ponysitters Club"

    tv_name_terms = tv_name + " TV Series"
    tv_name_terms = tv_name_terms.replace(" ", "+")
    wiki_search_url = "https://en.wikipedia.org/wiki/Special:Search?search="+tv_name_terms+"&go=Go&ns0=1"

    res = requests.get(wiki_search_url)
    html_page = res.content
    soup = BeautifulSoup(html_page, "html.parser")
    content = soup.find_all("li", {"class": "mw-search-result"})

    no_wiki_page = False
    wiki_url = "N/A"

    tv_name_terms = tv_name.split(" ")

    for result in content:
        if "TV" in result.text and (tv_name_terms[0] in result.text or tv_name_terms[-1] in result.text):
            if ("List of" not in result.text):
                if "TV" in (str(result.a["href"])+"\n"):
                    wiki_url = "https://en.wikipedia.org" + str(result.a["href"])
        #             print(tv_name, wiki_url, sep="\t")
        #             print(tv_name_terms[0], tv_name_terms[1], wiki_url, sep="\t")
                    if (tv_name_terms[0] not in wiki_url) or (tv_name_terms[-1] not in wiki_url):
                        wiki_url = "N/A"



    based_on = "N/A"
    try:
        res = requests.get(wiki_url)
    except:
        wiki_url = "N/A"
    try:
        html_page = res.content
        soup = BeautifulSoup(html_page, "html.parser")
        content = soup.find("table", {"class": "infobox vevent"})
        content = content.find_all("tr")

        for result in content:
            if "Based on" in result.text:
                based_on = result.text.split("by")[1].strip()
    except:
        pass
    
    wiki_data = {
        "wiki_url":wiki_url,
        "based_on":based_on
    }
    return wiki_data
    #     print(tv_name)
    # print(no_wiki_page, wiki_url, based_on, sep="\n")
```

```python
def get_imdb_rt_wiki_details(tv_name = "Hemlock Grove"):
    #OMDB values
    result = client.get(title=tv_name, tomatoes=True)
    try:
        imdb_rating = result["imdb_rating"]
        imdb_votes = result["imdb_votes"]
    except:
        imdb_rating = "N/A"
        imdb_votes = "N/A"
    try:
        imdb_id = result["imdb_id"]
    except:
        imdb_id = "N/A"
    try:
        tomato_critics = result["tomato_meter"]
        tomato_users = result["tomato_user_meter"]
    except:
        tomato_critics = "N/A"
        tomato_users = "N/A"
    try:
        show_writer = result["writer"]
    except:
        show_writer = "N/A"

    #Wikipedia
    wiki_data = get_wiki_url(tv_name)
    wiki_url = wiki_data["wiki_url"]
    based_on = wiki_data["based_on"]
    
    #Rotten Tomatoes
    if "N/A" in tomato_critics or "N/A" in tomato_users:
        rt_tv_name = tv_name.replace(" ", "_")
        rt_url = "https://www.rottentomatoes.com/tv/" + rt_tv_name

        res = requests.get(rt_url)
        html_page = res.content
        soup = BeautifulSoup(html_page, "html.parser")
        try:
            critics_content = soup.find("div", {"class": "mop-ratings-wrap__half critic-score"})
            users_content = soup.find("div", {"class": "mop-ratings-wrap__half audience-score"})

            tomato_critics = (critics_content.text.strip().split("\n")[0])
            tomato_users = (users_content.text.strip().split("\n")[0]) 
        except:
            tomato_critics = "Wrong Url"
            tomato_users = "Wrong Url"   


    imdb = {
        "imdb_id":imdb_id, 
        "imdb_rating":imdb_rating, 
        "imdb_votes":imdb_votes
    }
    tomato = {
        "tomato_critics":tomato_critics, 
        "tomato_users":tomato_users
    }

    result = {
        "imdb":imdb,
        "tomato":tomato,
        "show_writer":show_writer,
        "wiki_url":wiki_url,
        "based_on":based_on
    }
    return (result)
```
