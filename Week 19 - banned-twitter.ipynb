```python
import tweepy
import datetime
import xlsxwriter
import sys

# credentials from https://apps.twitter.com/
consumerKey = API_KEY
consumerSecret = API_SECRET_KEY
accessToken = ACCESS_TOKEN
accessTokenSecret = ACCESS_TOKEN_SECRET

auth = tweepy.OAuthHandler(consumerKey, consumerSecret)
auth.set_access_token(accessToken, accessTokenSecret)

api = tweepy.API(auth)



def get_tweets(handle_name="GameOfThrones", start=datetime.datetime(2021, 1, 1, 0, 0, 0), end=datetime.datetime(2021, 1, 1, 0, 0, 0)):
    count = 0
    username = handle_name
    startDate = start
    endDate = end

    tweets = []
    tmpTweets = api.user_timeline(username)
    for tweet in tmpTweets:
        if count == 0:
            tweets.append(tweet)
        else:
            break
        count += 1

    workbook = xlsxwriter.Workbook("handles/"+username + str(startDate.date()) + ".xlsx")
    worksheet = workbook.add_worksheet()
    row = 0
    worksheet.write_string(row, 0, "Tweet ID")
    worksheet.write_string(row, 1, "Date of tweet")
    worksheet.write_string(row, 2, "Content")
    worksheet.write_string(row, 3, "No. of followers")
    worksheet.write_string(row, 4, "No. of following")
    worksheet.write_string(row, 5, "No. of tweets")
    worksheet.write_string(row, 6, "No. of likes")
    worksheet.write_string(row, 7, "No. of retweets")
    worksheet.write_string(row, 8, "Tweet in reply to")

    # Tweet Id = tweet.id
    # Tweet Text = tweet.text
    # Tweet Created On = tweet.created_at
    # No. of followers of the account = tweet.user.followers_count
    # No. of following = tweet.user.friends_count
    # No. of tweets made by an account = tweet.user.statuses_count
    # No. of likes = tweet.favorite_count
    # No. of retweets = tweet.retweet_count
    # Tweet in reply to = tweet.in_reply_to_status_id

    row+=1
    for tweet in tweets:
        worksheet.write_string(row, 0, str(tweet.id))
        worksheet.write_string(row, 1, str(tweet.created_at))
        worksheet.write_string(row, 2, str(tweet.text))
        worksheet.write_string(row, 3, str(tweet.user.followers_count))
        worksheet.write_string(row, 4, str(tweet.user.friends_count))
        worksheet.write_string(row, 5, str(tweet.user.statuses_count))
        worksheet.write_string(row, 6, str(tweet.favorite_count))
        worksheet.write_string(row, 7, str(tweet.retweet_count))
        worksheet.write_string(row, 8, str(tweet.in_reply_to_status_id))
        row += 1

    workbook.close()
    print("Excel file ready")
```

```python
import snscrape.modules.twitter as sntwitter
import pandas
import datetime

# Creating list to append tweet data to
for indx in df.index:
    tweets_list2 = []
    handle = df["handle name"].iloc[indx]
    since = df["formatteddate"].iloc[indx]
    
    if str(handle) == "nan":
        continue
    until = (datetime.datetime.strptime(since, '%Y-%m-%d') + datetime.timedelta(days=7)).strftime('%Y-%m-%d')
    # handle, since, until
    # # Using TwitterSearchScraper to scrape data and append tweets to list
    for i,tweet in enumerate(sntwitter.TwitterSearchScraper("from:" + handle + " since:"+
                                                            since+" until:"+until).get_items()):
        if i>1500:
            break
        tweets_list2.append([tweet.date, tweet.id, tweet.content])

    # Creating a dataframe from the tweets list above
    tweets_df2 = pd.DataFrame(tweets_list2, columns=['Datetime', 'Tweet Id', 'Text'])
    tweets_df2.to_csv("handles/"+tweet.username+"-"+since+".csv")
```

```python
import os
path = 'handles'
files = os.listdir(path)

handle_info = pd.DataFrame()

for file in files:
    df = pd.read_excel(path+"/"+file)
    try:
        row = {
            "handle_name":str(file).replace("2021-01-01.xlsx", ""),
            "latest_tweet_date":df["Date of tweet"][0],
            "latest_tweet_text":df["Content"][0],
            "latest_tweet_likes":df["No. of likes"][0],
            "latest_tweet_retweets":df["No. of retweets"][0],
            "num_followers":df["No. of followers"][0],
            "num_following":df["No. of following"][0]
        }
        handle_info = handle_info.append(row, ignore_index=True)
    except:
        print(file)
    
handle_info["latest_tweet_likes"] = handle_info["latest_tweet_likes"].astype(int)
handle_info["latest_tweet_retweets"] = handle_info["latest_tweet_retweets"].astype(int)
handle_info["num_followers"] = handle_info["num_followers"].astype(int)
handle_info["num_following"] = handle_info["num_following"].astype(int)

display(handle_info)
handle_info.to_excel("handle_info.xlsx")
```
